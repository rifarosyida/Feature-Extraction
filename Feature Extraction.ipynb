```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import TfidfTransformer 
import pandas as pd

# Import data sheet
file_data = open('orange.txt', 'r').read()
output = open('output_raw.txt', 'w') # output file dataset dengan angka dan tanda baca/karakter yang sudah dibersihkan 
output_final = open('output_final.txt', 'w') # file output_raw.txt dengan baris yang kosong yang sudah dibersihkan 
# Removing dots
file = file_data.split('.')

# digunakan untuk menampung semua teks (hanya alphabet dan spasi)
alphaandspace = " "

for x in file: # melakukan perulangan pada teks yang dipisahkan dengan titik
    a_string = x;
    for character in a_string: # melakukan perulangan perkarakter untuk menghilangkan karakter selain alphabet dan spasi
        if character.isalpha() or character.isspace():
            alphaandspace = alphaandspace + character # menggabungkan kata yang telah dicek karakternya

output.write(alphaandspace) # menulis hasil teks ke dalam file bernama output_raw.txt

emptyRow = ""
# untuk menghilangkan enter
with open("output_raw.txt","r") as f:
 for i in f.readlines():
       if not i.strip():
           continue
       if i:
           emptyRow += i

output_final.write(emptyRow) # menuliskan teks yang enternya hilang ke dalam file output_final.txt

final_file = open('output_final.txt', 'r').readlines() # membaca file output_final.txt
lists = [x.replace('\n', '')for x in final_file] # masukin final output dan ngilangin \n ke dalam list

# print(lists)

vectorizer = TfidfVectorizer(stop_words='english')
response = vectorizer.fit_transform(lists)
print(response) 
```

```python
features_name = vectorizer.get_feature_names()
features_name
```
